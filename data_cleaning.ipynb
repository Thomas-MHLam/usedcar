```python
#data cleaning
import pandas as pd
from dateutil import parser
from sklearn.impute import KNNImputer

#plot
import matplotlib.pyplot as plt
```

```python
#data read-in
df = pd.read_csv(r"C:\Users\lamma\Side Project\Used car price modelling\vehicles.csv")
gdp_data = pd.read_csv(r"C:\Users\lamma\Side Project\Used car price modelling\us_gdp.csv")
```

```python
df.drop(['id', 'region', 'url', 'region_url', 'cylinders', 'VIN', 'size', 'condition', 'image_url','transmission', 'fuel', 'description','paint_color','title_status', 'drive', 'type', 'county', 'lat', 'long'], axis=1, inplace=True)
gdp_data['year'] = pd.to_datetime(gdp_data['date']).dt.year
```

```python
gdp_data
```

```python
cleaned_df = df.dropna(subset=['year', 'manufacturer', 'odometer', 'posting_date', 'model'])
cleaned_df['year'] = cleaned_df['year'].astype(int)
cleaned_df['state'] = df['state'].str.upper()

# Parse the date strings in the column using dateutil.parser
cleaned_df['posting_date'] = cleaned_df['posting_date'].apply(lambda x: parser.parse(x))
cleaned_df['posting_year'] = pd.to_datetime(cleaned_df['posting_date'], format='%Y-%m-%d %H:%M:%S%z', utc=True).dt.year
cleaned_df.drop('posting_date', axis=1)
```

```python
cleaned_df
```

```python
#data grouping
west_coast = ['WA', 'OR', 'CA']
east_coast = ['ME', 'NH', 'VT', 'MA', 'RI', 'CT', 'NY', 'NJ', 'PA', 'DE', 'MD', 'VA', 'NC', 'SC', 'GA', 'FL']
gulf_coast = ['AL', 'MS', 'LA', 'TX']
midwest = ['OH', 'IN', 'IL', 'MI', 'WI', 'MN', 'IA', 'MO', 'ND', 'SD', 'NE', 'KS']
mountain = ['MT', 'ID', 'WY', 'NV', 'UT', 'CO', 'AZ', 'NM']
southwest = ['OK', 'TX']
northwest = ['AK']
hawaii = ['HI']
alaska = ['AK']

coast_mapping = {
    'West Coast': west_coast,
    'East Coast': east_coast,
    'Gulf Coast': gulf_coast,
    'Midwest': midwest,
    'Mountain': mountain,
    'Southwest': southwest,
    'Northwest': northwest,
    'Hawaii': hawaii,
    'Alaska': alaska
}

cleaned_df['Coast'] = cleaned_df['state'].apply(lambda x: next((coast for coast, states in coast_mapping.items() if x in states), 'Other'))
```

```python
cleaned_df = cleaned_df.drop(['posting_date'], axis=1)
```

```python
gdp_data['year'] = pd.to_datetime(gdp_data['date']).dt.year
```

```python
gdp_data
```

```python
gdp_data = gdp_data.drop(['date', ' GDP ( Billions of US $)', ' Per Capita (US $)'], axis=1)
```

```python
merged_data = pd.merge(cleaned_df, gdp_data, on='year', how='left')
```

```python
merged_data 
```

```python
cleaned_data = merged_data[(merged_data['price'] <= 100000) & (merged_data['price'] >= 1000) 
                           & (merged_data['year'] >= 2000) 
                           & (merged_data['odometer'] <= 500000) ]
```

```python
cleaned_data
```

```python
# Create the KNN Imputer object
imputer = KNNImputer(n_neighbors=582)

# Apply the imputation
cleaned_data['Annual % Change'] = imputer.fit_transform(cleaned_data[[' Annual % Change']])
```

```python
cleaned_data.drop([' Annual % Change'], axis=1)
```

```python
cleaned_data.to_csv('concat_data.csv')
```
